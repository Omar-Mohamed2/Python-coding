```python
# abstract class
from abc import ABC ,abstractmethod
class shape(ABC):
    @abstractmethod
    def area():
        pass
    
# all classes inherit from it must implement all abstract method
    # simple example
class rect(shape):
    def __init__(self,height=0,width=0):
        self.height=height
        self.width=width
    def area(self):
        area=self.height * self.width
        return area
    # example for static method
    @staticmethod
    def pi():
        pi=3.14
        return pi
    
rect1=rect(5,3)
print("rect1 area=",rect1.area())
```

```python
# dander function on previous example
# abstract class


from abc import ABC ,abstractmethod
class shape(ABC):
    @abstractmethod
    def area():
        pass
    
# all classes inherit from it must implement all abstract method
    # simple example


class rect(shape):
    def __init__(self,height=0,width=0):
        self.height=height
        self.width=width
    def area(self):
        area=self.height * self.width
        return area
    

    # example for static method
    @staticmethod
    def pi():
        pi=3.14
        return pi
    

    #dander fun
    def __add__(self,other):
        areas=self.area() +other.area()
        return f"sum of area for 2 rect={areas}"

    
rect1=rect(5,3)
print("rect1 area=",rect1.area())
print("_"*50)


rect2=rect(5,4)
print("rect2 area=",rect2.area())
print("_"*50)


print(rect1 + rect2)

print("_"*50)


# a="a"
# print(dir(a))
```

```python
class queue:
    num_ob=0
    def __init__(self):
        self.list=[]
        queue.num_ob+=1

    def insert(self,value):
        self.list.append(value)

    def pop(self):
        if not self.is_empty:
         self.list.pop(0)
        else:
           print("the queue is empty")

    def is_empty(self):
        def is_empty(self):
          return len(self.list) == 0
        
    def print_q(self):
       print(f"the new queue={self.list}")



class Ex(Exception):
    pass

class NamedQueue(queue):
    queus={}
    def __init__(self, name, size):
        super().__init__()
        self.name = name
        self.size = size
        NamedQueue.queus[name]=self


    def insert(self, value):
        try:
            if len(self.list) >= self.size:
                raise Ex("Cannot be added, queue is at maximum size")
        except Ex as e:
            print(e)
        else:
            super().insert(value)
    @staticmethod
    def get_queue_by_name(name):
        return NamedQueue.queus.get(name, None)

q1 = NamedQueue("example_queue", 3)
for i in range(4):
  q1.insert(i+1)
print("the first q=")
q1.print_q()
print("_"*40)
q2 = NamedQueue("example_queue2", 3)
for i in range(3):
  q2.insert(i+2)
print("the first q2=")
q2.print_q()
print("_"*40)
print(f"the second  q ={q2.get_queue_by_name('example_queue')}")
```
